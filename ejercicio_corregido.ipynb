```python
# Contenido de nodo.py
class Nodo:

    def __init__(self, nombre, duracion):
        self.nombre = nombre
        self.duracion = duracion
        self.izq = None
        self.der = None
```

```python
class ArbolBST:

    def __init__(self):
        self.raiz = None
    # ==========================
    # INSERTAR
    # ==========================
    def insertar(self, nombre, duracion):
        self.raiz = self.__insertar(self.raiz, nombre, duracion)
    def __insertar(self, nodo, nombre, duracion):
        if nodo is None:
            return Nodo(nombre, duracion)
        if duracion < nodo.duracion:
            nodo.izq = self.__insertar(nodo.izq, nombre, duracion)
        elif duracion > nodo.duracion:
            nodo.der = self.__insertar(nodo.der, nombre, duracion)
        return nodo

    # ==========================
    # BUSCAR
    # ==========================

    def buscar(self, duracion):
        return self.__buscar(self.raiz, duracion)
    def __buscar(self, nodo, duracion):
        if nodo is None:
            return None
        if duracion == nodo.duracion:
            return nodo
        if duracion < nodo.duracion:
            return self.__buscar(nodo.izq, duracion)
        return self.__buscar(nodo.der, duracion)

    # ==========================
    # RECORRIDO INORDEN
    # ==========================

    def inorden(self):
        self.__inorden(self.raiz)
        print("\n===== TIEMPO TOTAL =====")
        print(f"Tiempo total: {musica.calcular_tiempo_total()} segundos")

    def __inorden(self, nodo):
        if nodo:
            self.__inorden(nodo.izq)
            print(f"{nodo.nombre} - {nodo.duracion} segundos")
            self.__inorden(nodo.der)

    # ==========================
    # PREORDEN
    # ==========================
    def preorden(self):
        self.__preorden(self.raiz)
    def __preorden(self, nodo):
        if nodo:
            print(f"{nodo.nombre} - {nodo.duracion}")
            self.__preorden(nodo.izq)
            self.__preorden(nodo.der)
    # ==========================
    # POSTORDEN
    # ==========================
    def postorden(self):
        self.__postorden(self.raiz)
    def __postorden(self, nodo):
        if nodo:
            self.__postorden(nodo.izq)
            self.__postorden(nodo.der)
            print(f"{nodo.nombre} - {nodo.duracion}")
    # ==========================
    # MÍNIMO
    # ==========================
    def minimo(self, nodo):
        while nodo.izq:
            nodo = nodo.izq
        return nodo
    # ==========================
    # ELIMINAR
    # ==========================
    def eliminar(self, duracion):
        self.raiz = self.__eliminar(self.raiz, duracion)
    def __eliminar(self, nodo, duracion):
        if nodo is None:
            return nodo
        if duracion < nodo.duracion:
            nodo.izq = self.__eliminar(nodo.izq, duracion)
        elif duracion > nodo.duracion:
            nodo.der = self.__eliminar(nodo.der, duracion)
        else:
            if nodo.izq is None:
                return nodo.der
            elif nodo.der is None:
                return nodo.izq
            sucesor = self.minimo(nodo.der)
            nodo.nombre = sucesor.nombre
            nodo.duracion = sucesor.duracion
            nodo.der = self.__eliminar(
                nodo.der,
                sucesor.duracion
            )
    # ==========================
    # FILTRAR CANCIONES CORTAS
    # ==========================

    def filtrar_canciones_cortas(self, duracion_minima):
        self.raiz = self.__filtrar(self.raiz, duracion_minima)

    def __filtrar(self, nodo, duracion_minima):

        if nodo is None:
            return None

        nodo.izq = self.__filtrar(nodo.izq, duracion_minima)
        nodo.der = self.__filtrar(nodo.der, duracion_minima)

        if nodo.duracion < duracion_minima:
            return self.__eliminar(nodo, nodo.duracion)

        return nodo
    # ==========================
    # PLAYLIST INTELIGENTE
    # ==========================

    def playlist_inteligente(self, minimo, maximo):
        canciones = []
        self.__playlist_inteligente(self.raiz, minimo, maximo, canciones)
        return canciones

    def __playlist_inteligente(self, nodo, minimo, maximo, canciones):

        if nodo is None:
            return

        if nodo.duracion > minimo:
            self.__playlist_inteligente(nodo.izq, minimo, maximo, canciones)

        if minimo <= nodo.duracion <= maximo:
            canciones.append((nodo.nombre, nodo.duracion))

        if nodo.duracion < maximo:
            self.__playlist_inteligente(nodo.der, minimo, maximo, canciones)
    # ==========================
    # RECOMENDAR CANCIÓN PERFECTA
    # ==========================

    def recomendar_cancion_perfecta(self, segundos):
        return self.__recomendar(self.raiz, segundos, None)

    def __recomendar(self, nodo, segundos, mejor):

        if nodo is None:
            return mejor

        if mejor is None or abs(nodo.duracion - segundos) < abs(mejor.duracion - segundos):
            mejor = nodo

        if segundos == nodo.duracion:
            return nodo

        if segundos < nodo.duracion:
            return self.__recomendar(nodo.izq, segundos, mejor)

        return self.__recomendar(nodo.der, segundos, mejor)
    
    # ==========================
    # MOSTRAR ÁRBOL
    # ==========================

    def mostrar_arbol(self):
        self.__mostrar_arbol(self.raiz, "", True)

    def __mostrar_arbol(self, nodo, prefijo, es_ultimo):

        if nodo is None:
            return

        print(prefijo, end="")

        if es_ultimo:
            print("└── ", end="")
            nuevo_prefijo = prefijo + "    "
        else:
            print("├── ", end="")
            nuevo_prefijo = prefijo + "│   "

        print(f"{nodo.nombre} ({nodo.duracion} s)")

        hijos = []

        if nodo.izq:
            hijos.append(nodo.izq)

        if nodo.der:
            hijos.append(nodo.der)

        for i, hijo in enumerate(hijos):
            self.__mostrar_arbol(
                hijo,
                nuevo_prefijo,
                i == len(hijos) - 1
            )
    # ==========================
    # TIEMPO TOTAL DE LA PLAYLIST
    # ==========================
    def calcular_tiempo_total(self):
        return self.__calcular_tiempo_total(self.raiz)

    def __calcular_tiempo_total(self, nodo):

        if nodo is None:
            return 0

        return (
            nodo.duracion
            + self.__calcular_tiempo_total(nodo.izq)
            + self.__calcular_tiempo_total(nodo.der)
        )
```

```python
# Crear el árbol de canciones
musica = ArbolBST()

musica.insertar("Radioactive", 186)
musica.insertar("Imagine", 187)
musica.insertar("Believer", 204)
musica.insertar("Shape of You", 233)
musica.insertar("Viva La Vida", 242)
musica.insertar("Counting Stars", 257)
musica.insertar("Perfect", 263)
musica.insertar("Someone Like You", 285)
```

```python
def menu():

    while True:

        print("\n" + "="*55)
        print(" SISTEMA DE RECOMENDACIÓN MUSICAL (BST)")
        print("="*55)
        print("1. Mostrar canciones")
        print("2. Buscar canción")
        print("3. Calcular tiempo total")
        print("4. Recomendar canción")
        print("5. Filtrar canciones cortas")
        print("6. Playlist Inteligente")
        print("7. Mostrar árbol BST")
        print("8. Eliminar canción")
        print("9. Salir")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "1":

            print("\n===== CANCIONES =====")
            musica.inorden()

        elif opcion == "2":

            duracion = int(input("Ingrese la duración de la canción: "))

            resultado = musica.buscar(duracion)

            if resultado:
                print("\nCanción encontrada")
                print(f"Nombre: {resultado.nombre}")
                print(f"Duración: {resultado.duracion} segundos")
            else:
                print("\nNo existe una canción con esa duración.")

        elif opcion == "3":

            total = musica.calcular_tiempo_total()

            minutos = total // 60
            segundos = total % 60

            print(f"\nTiempo total: {total} segundos")
            print(f"Equivale a {minutos} minutos y {segundos} segundos")

        elif opcion == "4":

            segundos = int(input("Duración deseada: "))

            recomendada = musica.recomendar_cancion_perfecta(segundos)

            print("\nCanción recomendada")
            print(f"Nombre: {recomendada.nombre}")
            print(f"Duración: {recomendada.duracion} segundos")

        elif opcion == "5":

            minimo = int(input("Eliminar canciones menores a: "))

            musica.filtrar_canciones_cortas(minimo)

            print("\nÁrbol actualizado correctamente.")

        elif opcion == "6":

            minimo = int(input("Duración mínima: "))
            maximo = int(input("Duración máxima: "))

            playlist = musica.playlist_inteligente(minimo, maximo)

            if len(playlist) == 0:
                print("\nNo existen canciones en ese rango.")

            else:

                print("\nPlaylist Inteligente\n")

                for nombre, duracion in playlist:

                    print(f"{nombre} - {duracion} segundos")

        elif opcion == "7":

            print("\n===== ÁRBOL BST =====\n")
            musica.mostrar_arbol()

        elif opcion == "8":

            duracion = int(input("Duración de la canción a eliminar: "))

            musica.eliminar(duracion)

            print("\nCanción eliminada correctamente.")

        elif opcion == "9":

            print("\nGracias por utilizar el sistema.")

            break

        else:

            print("\nOpción incorrecta.")
```
